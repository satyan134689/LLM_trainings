```python
import os
import requests
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from google.colab import drive
from huggingface_hub import login
from google.colab import userdata
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig
import torch
```

```python
deepseek = 'deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B'
```

```python
hf_token = userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
```

```python
system_message = """You are an Agent who generates Sythetic Data for shipping comapany to train there Machine Learning models for Anolmaly Detection,
in csv format without code blocks."""

user_prompt  = """Give me dataset for my Shipping company,
It should contain 15 columns for different moving parts of ships,
another 8 columns for external conditions of forces acting on ships,
Dataset should have 150 rows and dataset should be in csv format"""
```

```python

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(deepseek)
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(deepseek, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

## Define Data Generation Function

### Subtask:
Create a Python function that encapsulates the logic for generating the synthetic dataset. This function will take a user prompt as input, construct the appropriate messages for the language model, call the `model.generate` method using the pre-loaded `tokenizer` and `model`, and return the generated CSV data as a string.


**Reasoning**:
Define the `generate_csv_data` function as instructed, which will construct messages, tokenize them, generate model output, decode it, and extract the CSV data.



```python
def generate_csv_data(user_input_prompt: str) -> str:
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_input_prompt}
    ]
    inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
    outputs = model.generate(inputs, max_new_tokens=2000, pad_token_id=tokenizer.eos_token_id)
    decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=False)

    assistant_tag = "<|start_header_id|>assistant<|end_header_id|>\n\n"
    if assistant_tag in decoded_output:
        start_index = decoded_output.find(assistant_tag) + len(assistant_tag)
        csv_data = decoded_output[start_index:].strip()

        if "<|eot_id|>" in csv_data:
            csv_data = csv_data.split("<|eot_id|>")[0].strip()

        return csv_data
    else:
        return "Error: Could not find assistant's response."

print("Function `generate_csv_data` defined.")
```

**Reasoning**:
The function `generate_csv_data` has been defined. Now, create and launch the `gradio.Interface` using this function to provide a web-based interface for generating CSV data based on user input, as specified in the main task.



```python
import gradio as gr
iface = gr.Interface(
    fn=generate_csv_data,
    inputs=gr.Textbox(lines=5, label="Enter your prompt for CSV data generation:", placeholder="e.g., Give me a dataset with 10 columns and 50 rows about product sales."),
    outputs=gr.Textbox(lines=10, label="Generated CSV Data"),
    title="Synthetic CSV Data Generator for Shipping Company",
    description="Enter a prompt to generate synthetic CSV data for anomaly detection models in a shipping company context."
)

iface.launch(debug=True, share=True);
print("Gradio interface launched.")
```
